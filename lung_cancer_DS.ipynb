```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing data
df = pd.read_csv("lung cancer data.csv")
df.head(5)
```

```python
#calculate the percentage who has cancer between woman and men
df.groupby('GENDER')['LUNG_CANCER'].value_counts()
cancer_percentage = df[df['LUNG_CANCER']=="YES"]['GENDER'].value_counts()
```

```python
#Graphic of cancer distribution
labels=cancer_percentage.index
sizes= cancer_percentage.values

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.axis('equal')

plt.xlabel('Distribution of smokers per gender')


plt.show()
```

```python
bins =  [18,30,31,70,120]
label = ['18-30', '31-50', '50-70', '71+']
df['Age_Range']=pd.cut(df['AGE'], bins=bins, labels=label, right=True)
df.head(5)
```

```python
#Years groups suceptibility
cancer_per_year=df[df['LUNG_CANCER']=='YES']['Age_Range'].value_counts()
print(cancer_per_year)
```

```python
labels_year_range = cancer_per_year.index
sizes_year_range = cancer_per_year.values

plt.bar(labels_year_range,sizes_year_range,label="Bars 1",color="black")


plt.title("Age distribution")
plt.show()
```
